# 1. Populating a dictionary
Create a dictionary by using all the given variables.

```python
first_name = 'John'
last_name = 'Doe'
favorite_hobby = 'Python'
sports_hobby = 'gym'
age = 82
```

```python
my_dict = {
    'name': f'{first_name} {last_name}',
    'age': age,
    'hobbies': [favorite_hobby, sports_hobby]
}
```

```python
assert my_dict == {
        'name': 'John Doe',
        'age': 82,
        'hobbies': ['Python', 'gym']
    }
```

# 2. Accessing and merging dictionaries
Combine `dict1`, `dict2`, and `dict3` into `my_dict`. In addition, get the value of `special_key` from `my_dict` into a `special_value` variable. Note that original dictionaries should stay untouched and `special_key` should be removed from `my_dict`.

```python
dict1 = dict(key1='This is not that hard', key2='Python is still cool')
dict2 = {'key1': 123, 'special_key': 'secret'}
# This is also a away to initialize a dict (list of tuples)
dict3 = dict([('key2', 456), ('keyX', 'X')])
```

```python
# 'Your impelementation'
my_dict = {**dict1, **dict2, **dict3}
special_value = my_dict.pop('special_key')
```

The first line my_dict = {**dict1, **dict2, **dict3} merges three dictionaries: dict1, dict2, and dict3 into a new dictionary called my_dict. If there are any duplicate keys, the value from the dictionary appearing later in the merge operation will overwrite previous values. The second line special_value = my_dict.pop('special_key') removes the key-value pair associated with 'special_key' from my_dict and stores the value of 'special_key' into a new variable called special_value.

```python
assert my_dict == {'key1': 123, 'key2': 456, 'keyX': 'X'}
assert special_value == 'secret'

# Let's check that the originals are untouched
assert dict1 == {
        'key1': 'This is not that hard',
        'key2': 'Python is still cool'
    }
assert dict2 == {'key1': 123, 'special_key': 'secret'}
assert dict3 == {'key2': 456, 'keyX': 'X'}
```

If you try to access a key that doesn't exist, like portfolio['ROI'], Python will crash with a KeyError. Based on the "Initialization" section of the dictionary notebook, what is a safer method we can use to look up a key without risking a crash?

```python
# Example using .get() to safely access a key

# Assuming 'my_dict' is the dictionary we were working with
# my_dict = {'key1': 123, 'key2': 456, 'keyX': 'X'}

existing_key_value = my_dict.get('key1')
print(f"Value for 'key1': {existing_key_value}")

non_existing_key_value = my_dict.get('non_existent_key')
print(f"Value for 'non_existent_key' (default None): {non_existing_key_value}")

non_existing_key_with_default = my_dict.get('another_key', 'Default Value')
print(f"Value for 'another_key' (with custom default): {non_existing_key_with_default}")
```

# Practical Task: The Portfolio Update

```python
holdings_data = [
    {'ticker': 'AAPL', 'price': 170.00, 'shares': 10},
    {'ticker': 'GOOGL', 'price': 150.00, 'shares': 5}
]
display(holdings_data)
```

```python
first_stock = holdings_data[0]
first_stock_value = first_stock['price'] * first_stock['shares']
display(first_stock_value)
```

if we wanted to print a message only if first_stock_value is greater than 1000, what comparison operator would we use?

```python
if first_stock_value > 1000:
    print(f"The first stock's value ({first_stock_value}) is greater than 1000.")
else:
    print(f"The first stock's value ({first_stock_value}) is not greater than 1000.")
```
