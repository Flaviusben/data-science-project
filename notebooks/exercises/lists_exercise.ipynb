# 2. Create a new list without modifiying the original one


```python
original = ['I', 'am', 'learning', 'hacking', 'in']
```

```python
# Your implementation here
modified = original[:] #  creates a shallow copy of the original list
modified[3] = 'lists' # replaces the element at index 3 in the modified list (which was 'hacking') with the string 'lists'
modified.append('Python')
```

In Python, when you see [:] used with a list, like original[:], it's a way to create a shallow copy of the entire list. Essentially, it means "from the beginning of the list to the end of the list". This is a common and efficient way to create a new list that contains all the same elements as the original, without having the new list be a reference to the original. This ensures that any modifications to the new list (modified in your example) won't affect the original list.

```python
assert original == ['I', 'am', 'learning', 'hacking', 'in']
assert modified == ['I', 'am', 'learning', 'lists', 'in', 'Python']
```

# 3. Create a merged sorted list

```python
list1 = [6, 12, 5]
list2 = [6.2, 0, 14, 1]
list3 = [0.9]
```

```python
# Your implementation here
my_list = sorted(list1 + list2 + list3, reverse=True)
```

```python
print(my_list)
assert my_list == [14, 12, 6.2, 6, 5, 1, 0.9, 0]
```
