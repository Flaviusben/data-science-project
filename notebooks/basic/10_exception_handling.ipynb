# Exception Handling

Exceptions are errors that occur during program execution. Python provides a robust way to handle these errors gracefully using `try`, `except`, `else`, and `finally` blocks.

**Why handle exceptions?**
- Prevent program crashes
- Provide meaningful error messages
- Clean up resources (files, connections)
- Implement retry logic

## Basic `try-except` Block

The simplest form catches any exception that occurs.

```python
# Without exception handling - this would crash
# result = 10 / 0  # ZeroDivisionError

# With exception handling
try:
    result = 10 / 0
except:
    print('An error occurred!')
    result = None

print(f'Result: {result}')
```

## Catching Specific Exceptions

It's better practice to catch specific exceptions rather than all exceptions.

```python
def divide(a, b):
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print('Error: Cannot divide by zero!')
        return None

print(divide(10, 2))   # 5.0
print(divide(10, 0))   # Error message and None
```

## Multiple Exception Types

Handle different exceptions differently.

```python
def safe_list_access(lst, index):
    try:
        return lst[index]
    except IndexError:
        print(f'Error: Index {index} is out of range!')
        return None
    except TypeError:
        print('Error: Invalid index type!')
        return None

my_list = [1, 2, 3]
print(safe_list_access(my_list, 1))      # 2
print(safe_list_access(my_list, 10))     # IndexError
print(safe_list_access(my_list, 'abc'))  # TypeError
```

## Accessing Exception Details

You can access the exception object to get more information.

```python
def convert_to_int(value):
    try:
        return int(value)
    except ValueError as e:
        print(f'Conversion error: {e}')
        return None

print(convert_to_int('123'))      # 123
print(convert_to_int('abc'))      # Error with details
print(convert_to_int('12.5'))     # Error with details
```

## The `else` Clause

Code in the `else` block runs only if no exception was raised.

```python
def read_number(prompt):
    while True:
        try:
            value = input(prompt)
            number = int(value)
        except ValueError:
            print('Invalid input! Please enter a number.')
        else:
            # This runs only if conversion succeeded
            print(f'Successfully converted: {number}')
            return number

# Uncomment to test interactively:
result = read_number('Enter a number: ')
```

## The `finally` Clause

Code in the `finally` block **always** runs, whether an exception occurred or not. Perfect for cleanup operations.

```python
def read_file_safe(filename):
    file = None
    try:
        file = open(filename, 'r')
        content = file.read()
        return content
    except FileNotFoundError:
        print(f'Error: File {filename} not found!')
        return None
    except IOError as e:
        print(f'Error reading file: {e}')
        return None
    finally:
        # This always runs, ensuring file is closed
        if file:
            file.close()
            print('File closed.')

# Test with non-existent file
content = read_file_safe('nonexistent.txt')
```

## Raising Exceptions

You can raise exceptions manually using the `raise` keyword.

```python
def validate_age(age):
    if not isinstance(age, int):
        raise TypeError('Age must be an integer!')
    if age < 0:
        raise ValueError('Age cannot be negative!')
    if age > 150:
        raise ValueError('Age seems unrealistic!')
    return True

# Test with valid age
try:
    validate_age(25)
    print('Age is valid!')
except (TypeError, ValueError) as e:
    print(f'Validation failed: {e}')

# Test with invalid age
try:
    validate_age(-5)
    print('Age is valid!')
except (TypeError, ValueError) as e:
    print(f'Validation failed: {e}')
```

## Common Built-in Exceptions

- **`ValueError`**: Raised when a function receives an argument of the correct type but inappropriate value
- **`TypeError`**: Raised when an operation is applied to an object of inappropriate type
- **`KeyError`**: Raised when a dictionary key is not found
- **`IndexError`**: Raised when a sequence index is out of range
- **`FileNotFoundError`**: Raised when a file cannot be found
- **`ZeroDivisionError`**: Raised when dividing by zero
- **`AttributeError`**: Raised when an attribute reference or assignment fails
- **`ImportError`**: Raised when an import statement fails

## Practical Example: Safe Dictionary Access

```python
def get_user_info(users, user_id):
    """Safely retrieve user information."""
    try:
        user = users[user_id]
        name = user['name']
        email = user['email']

    except KeyError as e:
        print(f'Missing key: {e}')
        return None

    except TypeError:
        print('Error: Invalid data structure!')
        return None

    else:
        return f'{name} ({email})'

    finally:
        print('Lookup complete.')

# Test data
users = {
    1: {'name': 'Alice', 'email': 'alice@example.com'},
    2: {'name': 'Bob', 'email': 'bob@example.com'}
}

print(get_user_info(users, 1))      # Success
print(get_user_info(users, 99))     # Missing user
```

## Best Practices

1. **Be specific**: Catch specific exceptions, not all exceptions
2. **Don't silence errors**: At minimum, log the error
3. **Use finally for cleanup**: Close files, connections, etc.
4. **Fail fast**: Raise exceptions early when data is invalid
5. **Provide context**: Include helpful error messages
6. **Don't overuse**: Not every error needs try-except; sometimes letting it crash is better
